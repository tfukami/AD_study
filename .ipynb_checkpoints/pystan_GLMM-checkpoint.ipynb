```python
import pystan
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean1 = 10
mean2 = -10
mean3 = 0
num1 = 200
num2 = 300
num3 = 500
```

```python
X = np.concatenate([
    np.random.normal(loc=mean1, scale=1, size=num1),
    np.random.normal(loc=mean2, scale=1, size=num2),
    np.random.normal(loc=mean3, scale=1, size=num3)
])

np.random.shuffle(X)
```

```python
X
```

```python
plt.hist(X, bins=100, color='red', histtype='step')
plt.title("sample data")
plt.show()
```

```python
N = X.shape[0]
k = 3

stan_data = {'N': N, 'k':k, 'X':X}

fit = pystan.stan(file='gmm_mcmc.stan', data=stan_data, iter=10000)

print('Sampling finished.')
```

```python
fit.plot()
```

```python
fit.summary()
```
