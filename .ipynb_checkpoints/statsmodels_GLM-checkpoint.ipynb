```python
import statsmodels.api as sm
import pandas as pd
```

```python
data = pd.read_csv("data3a.csv")
```

```python
data
```

```python
x = sm.add_constant(data.x)
y = data.y
x, y
```

```python
model = sm.GLM(y, x, family=sm.families.Poisson())
res = model.fit()
```

```python
print(res.summary())
```
