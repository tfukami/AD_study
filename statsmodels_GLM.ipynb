参考：  
[StatsModels Linear Regression](http://statsmodels.sourceforge.net/stable/regression.html)

```python
data = pd.read_csv("data3a.csv")
```

```python
data
```

```python
x_ = data.ix[:,['x', 'y']]
x_
```

```python
x = sm.add_constant(data.ix[:,['x', 'y']])
y = data.y
x, y
```

```python
model = sm.GLM(y, x_, family=sm.families.Poisson(link=sm.genmod.families.links.log))
res = model.fit()
```

```python
print(res.summary())
```

```python
import patsy
from patsy import dmatrices
```

```python
data = data.reset_index()
```

```python
df = data.rename(columns={'level_0':'idx1', 'index':'idx2'})
df['f'][0]='A'
df
```

```python
y, X = dmatrices('idx1~f + x + y', data=df, return_type='dataframe')
```

```python
y, X
```

```python
mod = sm.OLS(y, X)
res = mod.fit()
```

```python
print(res.summary())
```
